```python
# 필요한 라이브러리 다시 임포트
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 준비 - BERT, ELECTRA, RoBERTa의 정확도만 포함
individual_models = ['BERT', 'ELECTRA', 'RoBERTa']
individual_accuracies = [0.64, 0.66, 0.70]

# 그래프 그리기 - 개별 모델 정확도만 표시
fig, ax = plt.subplots(figsize=(6, 6))

# 정확도 막대 그래프 그리기
sns.barplot(x=individual_models, y=individual_accuracies, palette='viridis', ax=ax)
ax.set_ylim(0.4, 0.8)
ax.set_ylabel('Accuracy')
ax.set_title('Individual Model Accuracies')

# 정확도 값 표시
for i, v in enumerate(individual_accuracies):
    ax.text(i, v + 0.01, f"{v:.2f}", ha='center', fontweight='bold')

plt.show()
```

```python
# 분류 보고서 데이터
classes = ['Negative', 'Positive']
precision = [0.76, 0.77]
recall = [0.78, 0.75]
f1_score = [0.77, 0.76]

x = np.arange(len(classes))  # 클래스 위치
width = 0.2  # 막대 너비

fig, ax = plt.subplots(figsize=(6, 6))

# 막대 그리기
rects1 = ax.bar(x - width, precision, width, label='Precision', color='lightgreen')
rects2 = ax.bar(x, recall, width, label='Recall', color='lightblue')
rects3 = ax.bar(x + width, f1_score, width, label='F1-Score', color='salmon')

# 레이블과 제목 설정
ax.set_ylabel('Scores')
ax.set_title('Weighted Voting')
ax.set_xticks(x)
ax.set_xticklabels(classes)
ax.set_ylim(0, 1)
ax.legend()

# 값 표시
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.show()
```

```python
# 데이터 준비 - Accuracy와 F1-score만 포함
models = ['BERT', 'RoBERTa', 'ELECTRA', 'Weighted Voting']
accuracies = [0.64, 0.66, 0.70, 0.76]
f1_scores = [0.64, 0.66, 0.70, 0.76]

# 데이터프레임 생성
import pandas as pd

data = {
    'Model': models,
    'Accuracy': accuracies,
    'F1-Score': f1_scores
}
df = pd.DataFrame(data)

# 그래프 그리기 - 정확도와 F1-Score 비교
fig, ax = plt.subplots(figsize=(10, 6))

# 정확도와 F1-Score를 비교하는 막대 그래프 그리기
x = np.arange(len(models))  # 모델 위치
width = 0.25  # 막대 너비

rects1 = ax.bar(x - width / 2, df['Accuracy'], width, label='Accuracy', color='skyblue')
rects2 = ax.bar(x + width / 2, df['F1-Score'], width, label='F1-Score', color='salmon')

# 레이블과 제목 설정
ax.set_ylabel('Scores')
ax.set_title('Comparison of Accuracy and F1-Score')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylim(0.4, 0.8)
ax.legend()

# 값 표시
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()
```

```python
# 데이터 준비 - 각 모델의 Positive와 Negative에 대한 F1-Score 포함
positive_f1_scores = [0.62, 0.66, 0.70, 0.76]
negative_f1_scores = [0.65, 0.65, 0.70, 0.77]

# 그래프 그리기 - Positive와 Negative 클래스에 대한 F1-Score 비교
fig, ax = plt.subplots(figsize=(12, 8))

# 모델 위치 및 막대 너비 설정
x = np.arange(len(models))  # 모델 위치
width = 0.20  # 막대 너비

# Positive와 Negative F1-Score에 대한 막대 그래프 그리기
rects1 = ax.bar(x - width, negative_f1_scores, width, label='Negative F1-Score', color='lightblue')
rects2 = ax.bar(x, positive_f1_scores, width, label='Positive F1-Score', color='lightgreen')
rects3 = ax.bar(x + width, df['F1-Score'], width, label='Overall F1-Score', color='salmon')

# 레이블과 제목 설정
ax.set_ylabel('F1-Score')
ax.set_title('Comparison of F1-Scores for Positive, Negative, and Overall for Different Models')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylim(0.4, 0.8)
ax.legend()

# 값 표시 함수
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# 각 막대에 값 표시
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.show()
```
