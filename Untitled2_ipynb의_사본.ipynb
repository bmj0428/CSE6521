```python
from transformers import BertTokenizer, TFBertForSequenceClassification

# BERT 토크나이저와 모델 가져오기 (bert-base-uncased 사용)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased')

# 모델 저장 (다음 사용을 위해 로컬 디렉토리에 저장)
model.save_pretrained('/content/models/bert_model')
tokenizer.save_pretrained('/content/models/bert_model')
```

```python
from datasets import load_dataset
import pandas as pd

# Hugging Face의 'yelp_polarity' 데이터셋을 로드합니다.
dataset = load_dataset('yelp_polarity', split='train')

# 데이터셋을 판다스 데이터프레임으로 변환
df = pd.DataFrame(dataset)

# 데이터프레임의 컬럼 확인
print(df.columns)

# 텍스트와 라벨 컬럼만 선택하여 새로운 데이터프레임으로 만듦
df = df[['text', 'label']]

# 데이터프레임에서 라벨이 정수로 되어 있는지 확인 (라벨 컬럼이 정수형이어야 함)
df['label'] = df['label'].astype(int)

# 헤더 없이 저장, 텍스트 필드를 큰따옴표로 감싸고 라벨은 큰따옴표 없이 저장
df.to_csv('/content/yelp_reviews_no_header.csv', index=False, header=False, quoting=1)

# 저장된 파일 일부 확인
with open('/content/yelp_reviews_no_header.csv', 'r') as f:
    for i, line in enumerate(f):
        if i < 5:  # 첫 5줄만 출력
            print(line)
```
